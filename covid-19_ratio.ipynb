```python
df.index.unique()
df['safe_line'] = np.array(0.02)
df.head(25)
```

```python
df['ratio'] = df['confirmed']/df['test']
dff1 = df[:12]
dff2 = df[12:41]
dff3 = df[41: 72]
dff4 = df[72: ]
dff2
```

```python
plt.style.use('ggplot')
fig, ax1 = plt.subplots(figsize = (15, 8))
ax2 = ax1.twinx()
line1 = ax1.plot(dff1.index, dff1['ratio'], color = 'r', linestyle = '--', marker = 'o', label = 'ratio')
line2 = ax1.plot(dff1.index, dff1['safe_line'], color = 'g', label = 'safe_line')
line3 = ax2.plot(dff1.index, dff1['confirmed'], color = 'b', label = 'confirmed')
ax1.set_title('ratio')
ax1.set_xlabel('date')
ax1.set_ylabel('ratio')
ax2.set_ylabel('population')
lines = line1 + line2 + line3
labels = [l.get_label() for l in lines]
plt.legend(lines, labels, loc = 2)
plt.grid(False)
plt.show()
```
